#### **1.Creating new dataframe to find the "GDP per capita" of some countries.**


```python
#Querying the dataframe to find the GDP per capita
df_gdp= df.query('indicator== "NY.GDP.PCAP.PP.CD"')
```

```python
df_gdp
```

```python
## Creating 2 new dataframe for the year "2016" and "2020" for visualizion.
df_gdp_2016 = df_gdp.query("Year==2016")
df_gdp_2020 = df_gdp.query("Year==2020")
```

```python
#Filtering the dataframe to get the info of the countries names India, United States, United Kingdom, Australia, China, Japan for the year 2016
df_gdp_2016 = df_gdp_2016[(df_gdp_2016['Country Name']=='India') | 
                          (df_gdp_2016['Country Name']=='United States') |
                          (df_gdp_2016['Country Name']=='Australia') | 
                          (df_gdp_2016['Country Name']=='United Kingdom') | 
                          (df_gdp_2016['Country Name']=='China') | 
                          (df_gdp_2016['Country Name']=='Japan')]
```

```python
df_gdp_2016
```

```python
#Plotting the bar chart for GDP per capita of the year 2016 for the countries 
fig = px.bar(
    df_gdp_2016,
    x= "Country Name",
    y=	"value",
    title ="Top 7 Countries by 2016 GDP Per Capita, PPP (current international $) ",
    labels = {"value" : "GDP Per Capita, PPP ( Current International $)"},
    color = "Country Name",
    template= "plotly_dark"
)
fig.show()
```

```python
#Filtering the dataframe to get the info of the countries names India, United States, United Kingdom, Australia, China, Japan for the year 2016

df_gdp_2020 = df_gdp_2020[(df_gdp_2020['Country Name']=='India') | 
                          (df_gdp_2020['Country Name']=='United States') | 
                          (df_gdp_2020['Country Name']=='Australia') | 
                          (df_gdp_2020['Country Name']=='United Kingdom') |
                          (df_gdp_2020['Country Name']=='China') |
                          (df_gdp_2020['Country Name']=='Japan')]
```

```python
df_gdp_2020
```

```python
#Plotting the bar chart for GDP per capita of the year 2016 for the countries
fig = px.bar(
    df_gdp_2020,
    x= "Country Name",
    y=	"value",
    title ="Top 7 Countries by 2020 GDP Per Capita, PPP (current international $) ",
    labels = {"value" : "GDP Per Capita, PPP ( Current International $)"},
    color = "Country Name",
    template= "plotly_dark"
)
fig.show()
```

#### **5.Creating new dataframe to find the "Carbon dioxide Depletion" of some countries.**

```python
#Finding the unique values in the indicator column
df["indicator"].unique()
```

```python
#Considering three unique values of the indicator column and listing them into a seperate variale
my_indicators = ["SP.POP.GROW", "SP.POP.TOTL", "NY.ADJ.DCO2.CD"]

#Filtering the data based on the list created
df1 = df[df["indicator"].isin(my_indicators)]
```

```python
df1
```

```python
#querying the data frame for the year 2016
df2 = df1.query('Year == 2016')
```

```python
df2
```

```python
#Filtering the dataframe to get the info of the countries names India, United States, United Kingdom, Australia, China, Japan for the year 2016


df_carbon_depletion_2016 = df2[(df2['Country Name']=='India') | 
                               (df2['Country Name']=='United States') | 
                               (df2['Country Name']=='Australia') | 
                               (df2['Country Name']=='United Kingdom') | 
                               (df2['Country Name']=='China') | 
                               (df2['Country Name']=='Japan')]
```

```python
df_carbon_depletion_2016
```

```python
#Creating a pivot table for better visualization

df_carbon_depletion_2016_pivot = pd.pivot_table(df_carbon_depletion_2016, 
                                                values = 'value', 
                                                index=['Country Code', 'Country Name','Year'], 
                                                columns = 'indicator').reset_index()

df_carbon_depletion_2016_pivot
```

```python
#Plotting the scatter plot for the year 2016
fig = px.scatter(
    df_carbon_depletion_2016_pivot,
    x="SP.POP.GROW",
    y="NY.ADJ.DCO2.CD", 
    title="Top 7 Countries by 2016 Adjusted savings: carbon dioxide depletion (current US$)", 
    color="Country Name", 
    template="plotly_dark",
    #size = "Country Name",
    #hover_name = "SP.POP.GROW"
)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="Population growth (annual %)",
    yaxis_title="Adjusted savings: carbon dioxide damage (current US$)",)


fig.show()
```
